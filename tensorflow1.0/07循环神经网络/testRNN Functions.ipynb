```python
import numpy as np
import tensorflow as tf
```

```python
x=np.random.normal(size=[10,5])
cls=3
X=tf.placeholder(tf.int32,[10,5])
X_oneHot=tf.one_hot(X,cls)
split=tf.split(X_oneHot,5,1)
squeezed=tf.squeeze(split[0],squeeze_dims=1)
```

```python
x
```

```python
with tf.Session() as sess:
    X_oneHot=sess.run(X_oneHot,feed_dict={X:x})
    split=sess.run(split,feed_dict={X:x})
    squeezed=sess.run(squeezed,feed_dict={X:x})
    print(squeezed)
    
```
