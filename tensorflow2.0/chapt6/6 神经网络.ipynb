### 6.3.1 张量的方式实现多层全连接网络

```python
# Initialize the parameters in each layer
W1=tf.Variable(tf.random.truncated_normal([784,256],stddev=0.1))
b1=tf.Variable(tf.zeros([256]))

W2=tf.Variable(tf.random.truncated_normal([256,128],stddev=0.1))
b2=tf.Variable(tf.zeros([128]))

W3=tf.Variable(tf.random.truncated_normal([128,64],stddev=0.1))
b3=tf.Variable(tf.zeros([64]))

W4=tf.Variable(tf.random.truncated_normal([64,10],stddev=0.1))
b4=tf.Variable(tf.zeros([10]))

X=tf.random.normal([2,784])
with tf.GradientTape() as tape:
    h1=X@W1+b1
    h1=tf.nn.relu(h1)
    h2=h1@W2+b2
    h2=tf.nn.relu(h2)
    h3=h2@W3+b3
    h3=tf.nn.relu(h3)
    h4=h3@W4+b4
```

```python
print(h4)
```

### 实例

```python
dataset_path=tensorflow.keras.utils.get_file('auto-mpg.data',
                                            'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data')
```

```python
# column names
import pandas as pd
column_names=['MPG','Cylinders','Displacement','Horsepower','Weight','Acceleration', 'Model Year', 'Origin']
raw_dataset=pd.read_csv(dataset_path,names=column_names,na_values='?',comment='\t',sep=' ',skipinitialspace=True)
```

```python
dataset=raw_dataset.copy()
# 删除有空数据的行
dataset=dataset.dropna()
# Origin表示产地，将其分离成3列
tmp=pd.get_dummies(dataset.Origin)
cols=['USA','Europe','Japan']
tmp.columns=cols
dataset=dataset.drop('Origin',axis=1)
dataset=pd.concat([dataset,tmp],axis=0)
dataset.fillna(0.0,inplace=True)
```

```python
# 分割原来数据集为训练集和测试集
train_dataset=dataset.sample(frac=0.8,random_state=42)
test_dataset=dataset.drop(train_dataset.index)
train_labels=train_dataset.pop('MPG')
test_labels=test_dataset.pop('MPG')
```

```python
dataset_stats=dataset.describe()
dataset_stats.pop('MPG')
dataset_stats=dataset_stats.transpose() # get the statistic information of the dataset
```

```python
#标准化数据
def norm(x):
    return (x-dataset_stats['mean'])/dataset_stats['std']

normed_train_data=norm(train_dataset)
normed_test_data=norm(test_dataset)
```

```python
# 生成tensors
train_db=tf.data.Dataset.from_tensor_slices((normed_train_data.values,train_labels.values))
train_db=train_db.shuffle(1000).batch(32)
```

```python
import seaborn as sns
cols=['Cylinders','Displacement','Weight','MPG']
sampleDataset=dataset[cols]
sns.pairplot(sampleDataset,diag_kind='kde')
```

```python
# 建立模型

from tensorflow.keras import layers,losses
from tensorflow import keras
class Model(keras.Model):
    def __init__(self):
        super(Model,self).__init__()
        self.fc1=layers.Dense(64,activation='relu')
        self.fc2=layers.Dense(64,activation='relu')
        self.fc3=layers.Dense(1)
    def call(self,inputs,training=None,mask=None):
        x=self.fc1(inputs)
        x=self.fc2(x)
        x=self.fc3(x)
        return x
```

```python
model=Model()
model.build(input_shape=(None,9))
model.summary()
```

```python
optimizer=tf.keras.optimizers.RMSprop(0.01)
```

```python
n_epochs=200
train_mae=[]
test_mae=[]
for epoch in range(n_epochs):

    for step,(x,y) in enumerate(train_db):
        with tf.GradientTape() as tape:
            out=model(x)
            loss=tf.reduce_mean(losses.MSE(out,y))
            mae_loss=tf.reduce_mean(losses.MAE(out,y))
            grads=tape.gradient(loss,model.trainable_variables)
            optimizer.apply_gradients(zip(grads,model.trainable_variables))
    train_mae.append(mae_loss)
    out=model(tf.constant(normed_test_data.values))
    test_mae.append(tf.reduce_mean(losses.MAE(test_labels.values,out)))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(train_mae)),train_mae,'b',label='train')
plt.plot(range(len(test_mae)),test_mae,'r',label='test')
plt.legend(loc='upper right')
```
